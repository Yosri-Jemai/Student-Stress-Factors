```python
#Data Display
r = 3
c = 7
it = 1
for i in data.columns:
    plt.subplot(r, c, it )#fix 
    if data[i].nunique() > 6:
        sns.kdeplot(data[i])
        plt.grid()
    else:
        sns.countplot(x=data[i])
    it += 1
plt.tight_layout()
plt.show()
```




*Which factor tend to report higher stress level?*

```python
# Calculate the correlation of every column
correlation_data= data.corr()

# Create a heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(data=correlation_data, annot=True, cmap='coolwarm')
plt.title("What tends to a higher stress level?", fontweight='bold')
plt.show()
```

As we can see, anxiety, depression, future career concerns, bullying, and headaches are highly correlated with increased stress levels. This indicates that students reporting higher levels of those factors are more likely to experience elevated stress levels.

Meanwhile, the heatmap reveals a distinct pattern for factors that appear to have a mitigating effect on stress. Students who report good self-esteem, strong social support networks, positive teacher-student relationships, and better sleep quality tend to reflect a lower stress levels.

Which physiological factor has the most impact on student stress?

```python
# Filtering the dataset with only physiological factors
physiological_factors = data[['headache', 'blood_pressure', 'sleep_quality', 'breathing_problem', 'stress_level']]

# Calculating correlation 
correlation_physiological = physiological_factors.corr()

# Visualizing heatmap with 
plt.figure(figsize=(10,8))
sns.heatmap(data=correlation_physiological, annot=True, cmap='coolwarm')
plt.title('Physiological factors vs Stress levels')
plt.show()
```

Among the physiological factors, headaches has the strongest correlation with stress levels, which means that students experiencing more frequent headaches are more likely to report higher stress levels.
Breathing problems and blood pressure also display positive correlations with stress, suggesting that these physiological factors can also contribute to stress levels in students.

On the other side, the data reveals a potentially reassuring trendâ€”students who report experiencing good sleep quality tend to exhibit lower stress levels. This association implies that ensuring students have better sleep quality may contribute to reducing their overall stress levels.
